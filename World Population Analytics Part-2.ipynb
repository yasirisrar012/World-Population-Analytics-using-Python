## Loading the Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world_population_data.csv')
```

```python
df
```

```python
df.head(20)
```

```python
df.tail(20)
```

# 2023 Population Distribution for Asia

```python
df_asia = df[df['continent']=='Asia']
df_asia
```

```python
df_asia.shape[0] # Total Number of Countries in Asia
```

```python
df_asia.columns
```

```python
df_asia.drop(columns=['rank','continent','2022 population', '2020 population', '2015 population',
       '2010 population', '2000 population', '1990 population',
       '1980 population', '1970 population', 'area (km²)', 'density (km²)',
       'growth rate', 'world percentage'],inplace=True)
df_asia
```

```python
total_asia_pop = df_asia['2023 population'].sum()
total_asia_pop
```

```python
df_asia['Percentage of Total Population'] = (df_asia['2023 population'] / total_asia_pop)*100
```

```python
df_asia
```

```python
labels = df_asia['country']
sizes = df_asia['Percentage of Total Population']
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
plt.title("Continent wise population distribution (2023) for Asia")
```

```python
df_asia = df_asia.head(20)
df_asia
```

```python
sns.barplot(x='2023 population',y='country',data=df_asia)
plt.title('Top 20 Countries of Asia with Highest Population')
plt.xlabel('Population (2023)')
plt.ylabel('Country')
plt.show()
```

## Correlation of Population Metrics for Different Continents

```python
df
```

### Asia

```python
df_asia = df[df['continent']=='Asia']
df_asia
```

```python
population_columns = ['2023 population', '2022 population', '2020 population',
                      '2015 population', '2010 population', '2000 population',
                      '1990 population', '1980 population', '1970 population']
correlation_matrix = df_asia[population_columns].corr()
plt.figure(figsize=(10, 8))  # Adjust the size for better readability
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Population Metrics for Asia')
plt.show()
```
