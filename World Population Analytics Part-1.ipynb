# 2023 Global Population Distribution by Continent 

```python
df_continent = df.groupby("continent")["2023 population"].sum().reset_index()
df_continent
```

```python
plt.plot(df_continent['continent'],df_continent['2023 population'],'o-r')
plt.bar(df_continent['continent'],df_continent['2023 population'])
plt.title("Total 2023 Population by Continent")
plt.xlabel("Continent")
plt.ylabel("Population (2023)")
```

```python
total_pop = df_continent['2023 population'].sum()
total_pop
```

```python
df_continent['Percentage of Total Population']=(df_continent['2023 population']/total_pop)*100
df_continent
```

```python
labels = df_continent['continent']
sizes = df_continent['Percentage of Total Population']
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
plt.title("Continent wise population distribution (2023)")
```

```python
sns.barplot(x="2023 population",y="country",data=top_20_countries)
plt.title("Top 20 Countries by Population in 2023")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```

## Correlation of Population Metrics

```python
population_columns = ['2023 population', '2022 population', '2020 population',
                      '2015 population', '2010 population', '2000 population',
                      '1990 population', '1980 population', '1970 population']
correlation_matrix = df[population_columns].corr()
plt.figure(figsize=(10, 8))  # Adjust the size for better readability
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Population Metrics')
plt.show()
```

## Population Growth over time

```python
population_columns = ['2023 population', '2022 population', '2020 population',
                      '2015 population', '2010 population', '2000 population',
                      '1990 population', '1980 population', '1970 population']
population_over_time = df.set_index('country')[population_columns]
population_over_time = population_over_time.sort_values(by='2023 population', ascending=False).head(20)
plt.figure(figsize=(15, 10))  # Adjust the size for better readability
sns.heatmap(population_over_time, cmap='YlGnBu', linewidths=.5, linecolor='gray',annot=True,fmt=".2f")
plt.title('Population Growth Over Time by Country')
plt.xlabel('Year')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.show()
```

## Density vs Area

```python
selected_columns = ['country', 'area (km2)', 'density (km2)']
df_selected = df[selected_columns]
plt.figure(figsize=(10, 8))  # Adjust the size for better readability
sns.heatmap(df_selected.set_index('country').head(20), cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5, linecolor='gray')
plt.title('Density vs Area Heatmap')
plt.xlabel('Metric')
plt.ylabel('Country')
plt.show()
```
